```python
total_positive_sentence = list()
with open ('./corpus.txt' , 'r' , encoding = 'utf-8') as rf:
    for s in rf.readlines():
        sentence = s.strip('\n')
        if "(" in sentence or "（" in sentence :
            continue
        elif '★' in sentence:
            continue
        elif '【' in sentence:
            continue
        elif '「' in sentence : 
            continue
        else:
            sentence_list = re.split(r'[;,？！《，。;；]\s*', sentence)
            total_positive_sentence += sentence_list
    
    print (len(total_positive_sentence))
```

```python
with open ('./positive_data.txt' , 'w' , encoding = 'utf-8') as wf:
    for i in total_positive_sentence:
        if len(i) < 7 : 
            continue
        else:
            wf.write(i + '\n')
```

```python
import csv
f = open ('./negative_data.txt' , 'w' , encoding ='utf-8' )
with open ('CJX_data.csv', 'r' , encoding = 'utf-8') as rf : 
    rows = csv.reader(rf) 
    for row in rows : 
        f.write(row[1].replace('{CJX}' , '') + '\n')
f.close()
    
```
