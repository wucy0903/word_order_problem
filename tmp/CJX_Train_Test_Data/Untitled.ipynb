```python
import monpa
import csv
```

```python
word_list = list()
tag_list = list()
cnt = 0
with open ('./CJX_data.csv','r',encoding='utf-8') as rf:
    csv_r = csv.reader(rf)
    for row in csv_r:
        cnt+=1
        print (cnt)
        cut_iter_list = list()
        tag_iter_list = list()
        cut_res = monpa.pseg(row[1])
        for res in cut_res:
            cut_iter_list.append(res[0])
            tag_iter_list.append(res[1])
        word_list.append(cut_iter_list)
        tag_list.append(tag_iter_list)
        print (cut_iter_list)
        print (tag_iter_list)
    
```

```python
list_sentence = list()
with open ('./CJX_data.csv','r',encoding='utf-8') as rf:
    csv_r = csv.reader(rf)
    for row in csv_r:
        list_sentence.append(row[1])
        
with open ('CJX_label_sentence.csv' , 'w' , encoding = 'utf-8') as wf:
    csv_w = csv.writer(wf)
    title_list = []
    cnt = 0
    for i in range (100):
        title_list.append(str(i))    
    csv_w.writerow(title_list)
    for s in word_list:
        labe_s = ['']
        labe_s = labe_s + ['0' for _ in range (len(s))]
        csv_w.writerow(labe_s)
        s = [''.join(s)]+s
        csv_w.writerow(s)
```

```python
total_tag_monpa = list()
with open ('tag_train.txt' , 'w', encoding = 'utf-8') as wf :
    for t in tag_list:
        wf.write(' '.join(t) + '\n')
        total_tag_monpa  = total_tag_monpa + t
        total_tag_monpa = list(set (total_tag_monpa))
with open ('../model_for_cjx/monpa_tag.txt' , 'w' , encoding = 'utf-8') as wf:
    for t in total_tag_monpa:
        wf.write(t + '\n')
        
```
