```python
def get_all_char():
    list_char = list()
    for i in range(0x4e00,0x9fa6):
        list_char.append(chr(i))
    return list_char
#print (get_all_char())
print (len(get_all_char()))
def file_name(file_dir):
    file = None
    for root, dirs, files in os.walk(file_dir): 
        #print(root) #當前目錄路徑 
        file = files
    return file
hsk_file_list = file_name('./hsk_data/')
#print (hsk_file_list)
def Simplified2Traditional(sentence):
    sentence = Converter('zh-hant').convert(sentence)
    return sentence
```

```python
def check_L (sentence) : 
    if '[L]' in sentence:
        while '[L]' in sentence:
            idx_s = sentence.index('L')-1
            idx_e = 100
            for t in range(idx_s,len(list(sentence))):
                if sentence[t] ==']':
                    idx_e = t
                    break
            sentence = sentence[:idx_s] + sentence[idx_e+1:]
    return sentence
def check_D(sentence):
    if '[D' in sentence :
        while '[D' in sentence:
            idx_s = sentence.index('D')-1
            idx_e = 100
            for t in range(idx_s,len(list(sentence))):
                if sentence[t] ==']':
                    idx_e = t
                    break
            sentence = sentence[:idx_s] + sentence[idx_e+1:]
    return sentence

def check_CC(sentence):
    if '{CC' not in sentence :
        return sentence
    else:
        idx = 0
        while idx != len(list(sentence)):
            #print ('idx : ' , idx)
            if '{CC' not in sentence :
                return sentence
            if '{' == sentence[idx]:
                #print ('check [',)
                if sentence[idx + 1] != 'C':
                    #print ('第一個不是C')
                    tp = idx +1
                    for t in range(tp,len(list(sentence))):
                        if sentence[t] == '}':
                            tp = t
                            idx = tp + 1
                            tp=0
                            break
                    continue
                else:
                    if sentence[idx + 2] !='C':
                        tp = idx +2
                        for t in range(tp,len(list(sentence))):
                            if sentence[t] == '}':
                                tp = t
                                idx = tp + 1
                                tp=0
                                break
                        continue
                    else:
                        tp = idx+2
                        for t in range(tp,len(list(sentence))):
                            if sentence[t] == '}':
                                tp = t
                                break
                        sentence = sentence[:idx] + sentence[tp+1:]
                        idx = idx
            idx += 1
        return sentence
    
def check_WWJ(sentence):
    if '{WWJ' not in sentence :
        return sentence
    else:
        idx = 0
        while idx != len(list(sentence)):
            #print ('idx : ' , idx)
            if '{WWJ' not in sentence :
                return sentence
            if '{' == sentence[idx]:
                #print ('check [',)
                if sentence[idx + 1] != 'W':
                    #print ('第一個不是C')
                    tp = idx +1
                    for t in range(tp,len(list(sentence))):
                        if sentence[t] == '}':
                            tp = t
                            idx = tp + 1
                            tp=0
                            break
                    continue
                else:
                    if sentence[idx + 2] !='W':
                        tp = idx +2
                        for t in range(tp,len(list(sentence))):
                            if sentence[t] == '}':
                                tp = t
                                idx = tp + 1
                                tp=0
                                break
                        continue
                    else:
                        tp = idx+2
                        for t in range(tp,len(list(sentence))):
                            if sentence[t] == '}':
                                tp = t
                                break
                        sentence = sentence[:idx] + sentence[tp+1:]
                        idx = idx
            idx += 1
        return sentence

def check_CQ(sentence):
    if '{CQ' not in sentence :
        return sentence
    else:
        idx = 0
        while idx != len(list(sentence)):
            #print ('idx : ' , idx)
            if '{CQ' not in sentence :
                return sentence
            if '{' == sentence[idx]:
                #print ('check [',)
                if sentence[idx + 1] != 'C':
                    #print ('第一個不是C')
                    tp = idx +1
                    for t in range(tp,len(list(sentence))):
                        if sentence[t] == '}':
                            tp = t
                            idx = tp + 1
                            tp=0
                            break
                    continue
                else:
                    if sentence[idx + 2] !='Q':
                        tp = idx +2
                        for t in range(tp,len(list(sentence))):
                            if sentence[t] == '}':
                                tp = t
                                idx = tp + 1
                                tp=0
                                break
                        continue
                    else:
                        tp = idx+2
                        for t in range(tp,len(list(sentence))):
                            if sentence[t] == '}':
                                tp = t
                                break
                        sentence = sentence[:idx] +sentence[idx+3:tp] + sentence[tp+1:]
                        idx = idx
            idx += 1
        return sentence
def check_CD(sentence):
    if '{CD' not in sentence :
        return sentence
    else:
        idx = 0
        while idx != len(list(sentence)):
            #print ('idx : ' , idx)
            if '{CD' not in sentence :
                return sentence
            if '{' == sentence[idx]:
                #print ('check [',)
                if sentence[idx + 1] != 'C':
                    #print ('第一個不是C')
                    tp = idx +1
                    for t in range(tp,len(list(sentence))):
                        if sentence[t] == '}':
                            tp = t
                            idx = tp + 1
                            tp=0
                            break
                    continue
                else:
                    if sentence[idx + 2] !='D':
                        tp = idx +2
                        for t in range(tp,len(list(sentence))):
                            if sentence[t] == '}':
                                tp = t
                                idx = tp + 1
                                tp=0
                                break
                        continue
                    else:
                        tp = idx+2
                        for t in range(tp,len(list(sentence))):
                            if sentence[t] == '}':
                                tp = t
                                break
                        sentence = sentence[:idx] + sentence[tp+1:]
                        idx = idx
            idx += 1
        return sentence
#要放在前面幾個check後面
def check_other(sentence):
    while '[' in list(sentence):
        idx_s = sentence.index('[')
        idx_e = 500
        for t in range(idx_s,len(list(sentence))):
            if sentence[t] == ']':
                idx_e = t
                break
        sentence = sentence[:idx_s] + sentence[idx_e+1:]
    return sentence
    
def check_p(sentence):
    #print (sentence)
    is_open = False
    if '（' in sentence:
        while '（' in sentence:
            idx_s = sentence.index('（')
            try:
                idx_e = sentence.index('）')
            except : 
                return 'Bad'
            if idx_s > idx_e:
                return 'Bad'
            sentence = sentence[:idx_s] + sentence[idx_e+1:]
    return sentence
```
